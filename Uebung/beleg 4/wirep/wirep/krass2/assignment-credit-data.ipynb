#### Description of the variables

- Target Variable `SeriousDlqin2yrs` [Y/N]: Person experienced 90 days past due delinquency or worse.

- `RevolvingUtilizationOfUnsecuredLines` [percentage]: Total balance on credit cards and personal lines of credit except real estate and no installment debt like car loans divided by the sum of credit limits.

- `DebtRatio` [percentage]: Monthly debt payments, alimony,living costs divided by monthy gross income
- `MonthlyIncome` [real]: Monthly income
- `NumberOfOpenCreditLinesAndLoans` [integer]: Number of Open loans (installment like car loan or mortgage) and Lines of credit (e.g. credit cards)
- `NumberOfTimes90DaysLate` [integer]: Number of times borrower has been 90 days or more past due.
- `NumberRealEstateLoansOrLines` [integer]: Number of mortgage and real estate loans including home equity lines of credit
- `NumberOfTime60-89DaysPastDueNotWorse` [integer]: Number of times borrower has been 60-89 days past due but no worse in the last 2 years.
- `NumberOfDependents` [integer]: Number of dependents in family excluding themselves (spouse, children etc.)

Cleaning Data
1. removing NaN
- either substitute NaN with zero, which lets you keep the data but makes it somewhat incorrect
- or remove all rows containing NaN which makes you lose data but leaves the remaining data in best shape

```python
print 'substitute NaN with 0.0'
df1 = df.fillna(0.0)
df1
```

```python
print 'ramoving rows containing NaN'
df2 = df.dropna()
df2
```

Note to self: need to reindex

```python
print 'now removing outliers'

upperThreashold = 0.98
lowerThresold = 0.02

#quantiles = df2.quantile([0.2, 0.8])
#print 'quantiles'
#print quantiles
mask = (df2 >= df2.quantile(lowerThresold)) & (df2 <= df2.quantile(upperThreashold))
#print 'mask:'
#print mask
df2 = df2[mask]
df3 = df2.dropna()
df3.describe()
```

Visualisation and interpretation of data

```python
#Age distribution

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(1,1,1)
ax.hist(df['age'], bins = 15, range = (df['age'].min(),df['age'].max()))
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.show()
```

```python
import seaborn as sns
sns.jointplot("age", "DebtRatio", df3, kind='reg')
```

```python
countTargetGroupedByAge = df3.groupby('age').SeriousDlqin2yrs.count()
probabilityTargetGroupedByAge = df3.groupby('age').SeriousDlqin2yrs.sum()/countTargetGroupedByAge

fig = plt.figure(figsize=(20,4))
ax1 = fig.add_subplot(121)
ax1.set_xlabel('age')
ax1.set_ylabel('number of people')
ax1.set_title("People by age")
countTargetGroupedByAge.plot(kind='bar')

ax2 = fig.add_subplot(122)
probabilityTargetGroupedByAge.plot(kind = 'bar')
ax2.set_xlabel('age')
ax2.set_ylabel('Probability of serious delinquency')
ax2.set_title("Probability of serious delinquency by age")
```

Age and dept ratio seem to have some degree of influence on the target variable as well as on each other.
